```python

drive_wheels_counts = df['drive-wheels'].value_counts().to_frame()
drive_wheels_counts.rename(columns={'drive-wheels': 'value_counts'}, inplace=True)
drive_wheels_counts
```

```python
drive_wheels_counts.index.name = 'drive-wheels'
drive_wheels_counts
```

```python
fuel_system=df['fuel-system'].value_counts().to_frame()
fuel_system.rename(columns={'fuel-system': 'value_counts'}, inplace=True)
fuel_system.index.name = 'fuel-system'
fuel_system.head(10) #especifica la cantidad 
```

```python
num_cilindros=df['num-of-cylinders'].value_counts().to_frame()
num_cilindros.rename(columns={'num-of-cylinders': 'value_counts'}, inplace=True)
num_cilindros.index.name = 'num-of-cylinders'
num_cilindros.head(12) #especifica la cantidad 
```
